```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
import matplotlib.image as mpimg
from zplane import zplane
```

```python
num = np.poly([0.8j,-0.8j])
den = np.poly([0.95*np.exp(1j*np.pi/8),0.95*np.exp(-1j*np.pi/8)])
Hz = num/den
zplane(num,den)
w, amp = signal.freqz(num,den)
angles = np.unwrap(np.angle(amp))
fig, ax1 = plt.subplots()
ax1.plot(w,20*np.log10(np.abs(amp)))
ax1.set_ylabel('Amplitude [dB]', color='b')
ax1.set_xlabel('Frquence normalis [rad/chantillon]')
ax2 = ax1.twinx()
ax2.plot(w, angles, 'g-')
ax2.set_ylabel('Phase (rad)', color='g')
ax2.grid(True)
ax2.axis('tight')
plt.show()
N= 500
impulse = [0 if i != N/2 else 1 for i in range(N)]
hn = signal.lfilter(num,den,impulse)
hn2 = signal.lfilter(den,num,hn)
plt.plot(hn)
plt.show()
plt.plot(hn2)
plt.show()


```

```python
#2
num = np.poly([np.exp(1j*np.pi/16),np.exp(-1j*np.pi/16)])
den = np.poly([0.95*np.exp(1j*np.pi/16),0.95*np.exp(-1j*np.pi/16)])
zplane(num,den)
w, amp = signal.freqz(num,den)
angles = np.unwrap(np.angle(amp))
fig, ax1 = plt.subplots()
ax1.plot(w,20*np.log10(np.abs(amp)))
ax1.set_ylabel('Amplitude [dB]', color='b')
ax1.set_xlabel('Frquence normalis [rad/chantillon]')
ax2 = ax1.twinx()
ax2.plot(w, angles, 'g-')
ax2.set_ylabel('Phase (rad)', color='g')
ax2.grid(True)
ax2.axis('tight')
plt.show()

N = 500
xn = [np.sin(n*np.pi/16)+np.sin(n*np.pi/32) for n in range(N)]
plt.plot(xn)
plt.show()
yn = signal.lfilter(num,den,xn)
plt.plot(yn)
plt.show()
```
